### Exercise 18.7

The core of a PWR contains 180 square fuel assemblies of length 4m, width 0.2m. 
- (a) Find the core volume and radius of an equivalent cylinder. 
- (b) If there are 200 fuel rods per assembly with pellets of diameter 0.9 cm, what is the approximate UO2 volume fraction of the core?

The equation for the volume of the assembly and the core, respectively, is given by:

$$V_a = l b h, V_c = 180 V_a $$

Then the area can be given by this equation, and we can rearrange to find R:

$$A = \pi R^2 $$

$$R = \sqrt{\frac{A}{\pi}} $$

The area of a pellet can be given by:

$$a = \pi R^2, where, R = \frac{d}{2}, so, a = \pi(\frac{d}{2})^2$$

```python
#This code is for part A
l = 4 #in meters
b = 0.2 #in meters
h = 0.2 #in meters
V_a = l*b*h
V_c = 180*V_a
print("(a) The volume of the equivalent cylinder is {:.1f}m".format(V_c))
#Because we know b and h, we can calculate the value of the area:
Area = 180*b*h
#Now that we have the area, we can calculate for R
R = math.sqrt(Area/math.pi)
print("(a) The radius of this equivalent cylinder is {:.2f}m".format(R))

#This code is for part B
d = 0.9 #given in question and in cm units
a_assembly = 200*math.pi*((d/2)*(d/2))
a_core = (180*a_assembly)/1e2
ratio = (a_core/Area)/1e2
print("(b) The approximate volume fraction of the core is {:.2f}".format(ratio))
```

### Exercise 18.9

Over the course of 3 years, a nuclear unit was shut down twice for refueling outages lasting
32 and 45 days, respectively, but was otherwise operating at full power. Determine the
capacity factor for this 3-year period.

The equation for the capacity factor is given by:

$$CF = \int_{0}^{T} \frac{P(t) \,dt}{P_e T}$$

$$CF = [\frac{(T-t) P_e}{T P_e}]$$

$$CF = \frac{(T-t)}{T}$$

```python
T = 3*365
t = 32+45
CF = (T-t)/T
print("The capacity factor for this 3-year period is {:.2f}".format(CF))
```

The number of fuel rods in a reactor can also be given by the equation:

$$N_R = \frac{Q_R}{q' H}$$

Where Q_R is the reactor's power output, q' is the power per unit length, and H is the core's height.

```python
#All numbers can be gotten from Table 18.4

#This code is for part A
Q_R = 3600*(1e3) #in MWt convert to kWt
q_prime = 5.4 #in Wt/ft
H = 145*(1/12) #in inches converted to feet
N_R = Q_R/(q_prime*H)
print("(a) The number of fuel rods in a PWR is {:.0f}".format(N_R))

#This code is for part B
#The same process as part A can be used.
Q_R = 3600*(1e3) #in MWt convert to kWt
q_prime = 6.1 #in Wt/ft
H = 148*(1/12) #in inches converted to feet
N_R = Q_R/(q_prime*H)
print("(b) The number of fuel rods in a BWR is {:.0f}".format(N_R))
```
