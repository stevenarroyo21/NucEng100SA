### Exercise 1.12
Show that to obtain a power of 1 W from  fission of uranium, it is necessary to cause $\sim3.3 \times 10^{10}$ fission events per second. Assume that each fission releases 190 MeV of useful energy.

```python
#We need to show that 1W of power is obtained from Uranium by causing 3.3x10^10 fission events per second.
#We know that each fission reaction releases 190 MeV of "useful" energy.

#First let's convert 190 MeV to eV and eV to Joules for an easier calculation.
MeV = 190
#MeV is 10^6 of an eV
eV = MeV*100000
#1 eV is equal to 1.6x10^-19 of a Joule
J = 1.6*0.000000000000000001
NewJ = eV*J
print("Now we know that 190MeV is equal to about {:.3e} J of energy.".format(NewJ))

#In order to calculate our fission events per second, we need to divide 1W by our New Joules Value
#1 W is equal to 1 Joule per second
W = 1
FissionEvents = W/NewJ
print("Thus, to obtain 1W from fission, we need {:.3e} fission events per second".format(FissionEvents))

#Rounded to the nearest tenth, our fission events necessary is around 3.3x10^10, which solves the problem.
```

### Exercise 3.4

### Exercise 3.7
There are approximately 140 g of potassium in an average person's body. From this weight, the abundance of potassium-40, and Avogadro's number, find the number of atoms. Find the decay constant in $s^{-1}$. How many disintegrations per second are there in the body? How many Bq and how many *micro*curies is that? 

```python
#First, we need to find the number of atoms N and the decay constant λ in s^-1.
#We can start out by finding the decay constant λ which is equal to 0.693/T_half where T_half is the half life.
#The half life value for potassium-40 is given in the book as 1.251x10^9 years.
λ = 0.693/((1.251*(10**9))*365*24*60*60)
print("The decay constant is {:.3e}.".format(λ))
#We now know that the decay constant λ is about 1.76x10^-17

#The greek letter γ is the unit we use for abundance. γ for potassium-40 is 0.00017 which is given in the book.
#The weight W is given by the mass times γ.
γ = 0.000117
m = 140
W = γ*m
print("The weight is equal to {:.3e} grams.".format(W))
#Now we know the weight W is equal to 0.01638g, and now we can find our # of atoms.
#The number of moles is equal to the weight W divided by the atomic mass A.
A = 39.0983
n = W/A
print("We have {:.3e} moles present.".format(n))

#Now we know that there is 4.19x10^-4 moles. Now we just have to multiply by Avogadro's number.
Avogadro = 6.023*100000000000000000000000
N = n*Avogadro
print("The correct number of atoms is {:.3e}.".format(N))
#The correct number of atoms in this potassium-40 sample is 2.523x10^20 atoms.

#Now for the next half of the problem, we need to find the rate of disintegration.
#The rate of disintegration is given by λ multiplied by the number of atoms N.
Disintegration = λ*N
print("There are around {:.3e} disintegrations per second.".format(Disintegration))
#Around 4432 disintegrations happen per second (dps).
#This value is in BECQUERELS

#Now we need to convert that dps from bequerels to curies.
#1 curie is 1/3.7x10^10 becquerels.
#However, we are looking for microcuries, so the conversion is now changed.
#1 microcurie is 1/3.7x10^4 becquerels
Becq = 3.7*10000
Disintegration_in_microCuries = Disintegration/Becq
print("This is the dps in microcuries: {:.3e}.".format(Disintegration_in_microCuries))
#The dps is 0.1198 microcuries.
```
